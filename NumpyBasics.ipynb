Basics


```python
import numpy as np
```

```python
arr_1D= np.array([1,2,3,4,5])
print("1D Array:", arr_1D)
```

```python
print("Multiplication by 2:", arr_1D * 2)
```

```python
print("zero matrix:", np.zeros((3, 4)))

print("one matrix:", np.ones((2, 5)))

print("Constant matrix:", np.full((3, 3), 7))

print("Random matrix:", np.random.rand(2, 3))

print("sequence array:", np.arange(0, 10))


```

#Properties Of Arrays

```python
arr= np.array([[1,2,3],[4,5,6]])
print("Array Properties:")
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
print("Size:", arr.size)
print("Number of Dimensions:", arr.ndim)
print("Transpose:\n", arr.T)

```
